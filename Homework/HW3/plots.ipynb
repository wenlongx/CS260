```python
# UNCOMMENT THE LINES BELOW TO GENERATE THE PLOTS USED IN THE REPORT

import matplotlib.pyplot as plt
import numpy as np
```

```python
mlp_loss = [1.88734,
            1.69665,
            1.60061,
            1.54682,
            1.50955,
            1.47790,
            1.45075,
            1.42686,
            1.41448,
            1.39606]

mlp_norelu_loss = [1.99805,
                   1.94767, 
                   1.92912,
                   1.92010,
                   1.91321,
                   1.91110,
                   1.90699,
                   1.90685,
                   1.90094,
                   1.89771]

cnn_loss = [1.70560,
            1.15904,
            1.06496,
            0.99840,
            0.94976,
            0.91520,
            0.88192,
            0.85888,
            0.82944,
            0.82432]

cnn_norelu_loss = [2.00152,
                   1.96289,
                   1.95506,
                   1.94734,
                   1.94991,
                   1.94522,
                   1.94947,
                   1.94479,
                   1.94229,
                   1.94586]
```

```python
losses = [
    ("MLP", mlp_loss, 54.58),
    ("MLP w/o ReLU", mlp_norelu_loss, 37.28),
    ("CNN", cnn_loss, 81.84),
    ("CNN w/o ReLU", cnn_norelu_loss, 36.63)
]
num_epochs = 10

plt.figure(figsize=(14,7))
for model_name, model_loss, acc in losses:
    plt.plot(np.linspace(1, num_epochs, num_epochs),                                         
             model_loss,
             label=f"{model_name}, Test Accuracy = {acc:.2f}%")

plt.title(f"Model Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.legend()
plt.show()
```

```python
mlp_losses = losses[0:2]
cnn_losses = losses[2:4]
num_epochs = 10

plt.figure(figsize=(6,6))
for model_name, model_loss, acc in mlp_losses:
    plt.plot(np.linspace(1, num_epochs, num_epochs),                                         
             model_loss,
             label=f"{model_name}, Test Accuracy = {acc:.2f}%")

plt.title(f"MLP Training Losses")
plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,6))
for model_name, model_loss, acc in cnn_losses:
    plt.plot(np.linspace(1, num_epochs, num_epochs),                                         
             model_loss,
             label=f"{model_name}, Test Accuracy = {acc:.2f}%")

plt.title(f"CNN Training Losses")
plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.legend()
plt.show()
```
